```python
def is_palindrome(n):
    rev=0;
    t=n
    while n>0:
        rev=rev*10+n%10
        n=n//10
    return rev==t
lst=[i for i in range(1,201) if is_palindrome(i)==True]
print(lst)
```

# sets in python
- Definition: Set is an unordered collection of unique immutable elements.
- Set will not hold duplicates.
- Set elements are enclosed within curly braces().
- Sets are mutable.
- Sets support insertion, deletion operations, just like lists.
- Sets are not substiptable (Sets are not indexed).
- You can only pass in immutable elements to set.
- Which means it's possible to create a set of strings, set of integers or a set of floating point values, but it's not
  possible to have a set of lists or even set of sets.

```python
s={10,20,30,40,10}
print(s)
print(type(s))
print(len(s))
print(max(s),min(s),sum(s))
```

# set()
- set() function is used to convert any other iterable into a set

```python
lst=[10,20,10,30,20,10,10,20,30,40]
print(len(lst))
s=set(lst)
print(s)
print(len(s))
```

# set.add()
- Used to add new elements to an existing set.
- set_name.add(new_element)

```python
s={10,20}
s.add(40)
s.add(10)
s.add(30)
s.add(20)
print(s)
```

## utility set methods
- if u have 2 sets s1 and s2
- union

```python
s1={10,20,30}
s2={30,40,50}
new_set_1=s1.union(s2)
print(new_set_1)
new_set_2=s1.intersection(s2)
print(new_set_2)
new_set_3=s1.difference(s2)
print(new_set_3)
new_set_4=s2.difference(s1)
print(new_set_4)
new_set_5=s1.symmetric_difference(s2)
print(new_set_5)
```
