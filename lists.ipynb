. On sequence types, you run two types of loops
1. Index Based Loops
2. Value Based Loops

# Index based loop

```python
lst=[10,20,30,40,50]
for i in range(len(lst)):
    print(lst[i],end=' ')
```

```python
lst=[10,20,30,40,50]
for i in range(len(lst),0,-1):
    print(lst[-i],end=' ')
for i in range(len(lst),0,-1):
    print(-i,end=' ')
```

# sort()
- Sorts the list in ascending order (in-place)
- Takes two option arguments - reverse, key
- When reverse is set to True, list gets sorted in descending order
- When given key, sort happens based on the return values of funtion call sent as key

```python
lst=[38,49,-20,40,28]
print(f"Before sorting: {lst}")
lst.sort()  # Tim sort (selection+Quick)
print(f"After sorting (Ascending): {lst}")
lst.sort(reverse=True)
print(f"Descending sorting (Descending): {lst}")
```

## key in list.sort()
- Key in list.sort() take a function
- And performs the sort based on the return value of the function

```python
names=['walter','skylerr','tuco','salamanca','jesse']
names.sort()
print(names)
```

```python
names=['walter','skylerr','tuco','salamanca','jesse']
names.sort(key=len)   #based on length of a char
print(names)
```

```python
def get_vowel_count(string):
    vc=0
    for i in string:
        if i in 'aeiou':
            vc+=1
        return vc
names=['walter','skylerr','tucoo','salamanca','jessiie']
names.sort(key=get_vowel_count)  ## reverse=True
print(names)  ## printing less no.of vowels wise in a word
```

```python
def get_a_count(string):
    vc=0
    for i in string:
        if i in 'a':
            vc+=1
    return vc
z=['babac','xan','aaabb','pbaacaaba']
z.sort(key=get_a_count)  ## reverse=True
print(z)
```

```python
def fac_count(n):
    count=0
    for i in range(1,n+1):
        if n%i==0:
            count+=1
    return count
nums=[12,6,9,17]
nums.sort(key=fac_count)
print(nums)
```

# copy()
- Return a shallow-copy of the existing list

```python
#Shallow Copy
nums=[10,20,30,40,50]
# Duplication of the list
num1=nums.copy() ## shallow-copy
print(num1)
nums.append(200)
nums.append(348)
print(nums)
print(num1)
```

```python
lst=['jesse','tuco','salamanca','walter','skyler']
print(lst)
lstt=lst.copy()
lst.sort()
print(lst)
cnt=0
for i in range(len(lst)):
    if lst[i]==lstt[i]:
        cnt+=1
print(cnt)
```

# List Comprehensions
- comphending a list is a easy and concise way to obtain other lists from existing lists(or any other iterables)
- List comprehensions often make it easy for a programmer to write code in as few lines as possible, allowing to obtain 
  same functionality that he/she would in more number of lines of code

```python
names=['jesse','tuco','salamanca','walter','skylerr']
new_list=[len(i) for i in names]
print(new_list)
new_list=[max(i) for i in names] #min
print(new_list)
```
