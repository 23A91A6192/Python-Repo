# filter(func,iterable)
- filter() takes a boolean function and an iterable.
- And filters in the values from the iterable that gives True when sent to the function
- filter() returns a filter object

```python
#filtering even numbers from a list of integers
def is_even(n):
    return n%2==0
nums=[2,4,9,8,16,17,31]
f=filter(is_even,nums)
print(list(f))
```

```python
#filtering primes from 1 to 100 using filter
def is_prime(n):
    cnt=0
    for i in range(1,n+1):
        if n%i==0:
            cnt+=1
    if cnt==2:
        return True
    else:
        return False
        
print(list(filter(is_prime,range(1,101))))
```

```python
def is_palindrome(n):
    temp=n
    rev=0
    while n!=0:
        r=n%10
        rev=rev*10+r
        n=n//10
    return temp==rev
f=filter(is_palindrome,range(1000,2000))
print(list(f))
```

# Inheritance in python
-  A class can inherit from multiple classes in python
-  Multiple Inheritance is supported in python
-  Syntax for inheritance
  class A(B)
- Here A is a child-class and B is a parent class
- If a parent class has a constructor without default values a child class object of that parent should be created in just the way parent class object is created.
- Otherwise, child-class constructor should make a call to super().__init__() to call it's parent constructor

```python
class parent:
    def method1(self):
        print("Hello")
    def method2(self,a,b):
        print(a+b)
class child(parent):
    pass
c1=child()
c1.method1()
c1.method2(10,20)
```

```python
class Mother:
    def operate(self,a,b):
        print(a+b)
class Father:
    def operate(self,a,b):
        print(a-b)
class Child(Father,Mother):
    pass
c1=Child()
c1.operate(10,20)
```

# Method overriding in python and polymorphism
- If a child-clas provides a specific implementation for a method that already exists in parent class, we call it as overriding
- In python, if both child and parent classes are having same method, the preferense is given to the method in child class first( if the method in child is not present it will look for that very method in parent class)

```python
class Parent:
    def say_hello(self):
        print("Hello, I am parent class")
class Child(Parent):
    def say_hello(self):
        print("Hello, I am child class")
c1=Child()
c1.say_hello()
```
